```python
from LxmlSoup import LxmlSoup
import requests
import pandas as pd
```

```python
html = requests.get('https://sunlight.net/catalog').text
html
```

```python
soup = LxmlSoup(html)
soup
```

```python
links = soup.find_all('a', class_='cl-item-link js-cl-item-link js-cl-item-root-link')
links
```

```python
url = list()
name = list()
price = list()


for i, link in enumerate(links):
    url.append(link.get("href"))
    name.append(link.text())
    price.append(soup.find_all("div", class_="cl-item-info-price-discount")[i].text())
    print(f"Url - {url}")
    print(f"Name - {name}")
    print(f"Price - {price}\n")
```

```python
d = {'url': url , 'name': name, 'price': price}
```

```python
df = pd.DataFrame(data=d)
df
```
